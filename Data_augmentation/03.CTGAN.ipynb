# Note: CTGAN usese Pytorch as default and not tensorflow

```python
import sdv
import pandas as pd
from sdv.single_table import CTGANSynthesizer
from sdv.metadata import SingleTableMetadata
print(sdv.version.public)
```

```python
base_path = "Data/Train_80/"
```

```python
# Data reading
healthy_tomato   = pd.read_excel(base_path + "tomato_healthy.xlsx")
unhealthy_tomato = pd.read_excel(base_path + "tomato_unhealthy.xlsx")
capsicum         = pd.read_excel(base_path + "capsicum.xlsx")
potato           = pd.read_excel(base_path + "potato.xlsx")
chilli           = pd.read_excel(base_path + "chilli.xlsx")
mulberry         = pd.read_excel(base_path + "mulberry.xlsx")

# Shapes
print("healthy_tomato:", healthy_tomato.shape)
print("Unhealthy_Tomato:", unhealthy_tomato.shape)
print("Capsicum:", capsicum.shape)
print("Potato:", potato.shape)
print("Chilli:", chilli.shape)
print("Mulberry:", mulberry.shape)
```

```python
# Converting labels as string and taking X and y

#healthy_Tomato
healthy_tomato.columns = healthy_tomato.columns.astype(str)
X_healthy_tomato = healthy_tomato.drop("label",axis = 1)
Y_healthy_tomato = healthy_tomato["label"]

#Unhealthy_Tomato
unhealthy_tomato.columns = unhealthy_tomato.columns.astype(str)
X_Unhealthy_Tomato = unhealthy_tomato.drop("label",axis = 1)
Y_Unhealthy_Tomato = unhealthy_tomato["label"]

#Capsicum
capsicum.columns = capsicum.columns.astype(str)
X_Capsicum = capsicum.drop("label",axis = 1)
Y_Capsicum = capsicum["label"]

#Potato
potato.columns = potato.columns.astype(str)
X_Potato = potato.drop("label",axis = 1)
Y_Potato = potato["label"]

#Chilli
chilli.columns = chilli.columns.astype(str)
X_chilli = chilli.drop("label",axis = 1)
Y_chilli = chilli["label"]

#Mulberry
mulberry.columns = mulberry.columns.astype(str)
X_Mulberry = mulberry.drop("label",axis = 1)
Y_Mulberry = mulberry["label"]
```

```python
# Metadata related

metadata1 = SingleTableMetadata()
metadata1.detect_from_dataframe(data=X_Unhealthy_Tomato)
# print(metadata1.to_dict())

metadata2 = SingleTableMetadata()
metadata2.detect_from_dataframe(data=X_Capsicum)
# print(metadata2.to_dict())

metadata3 = SingleTableMetadata()
metadata3.detect_from_dataframe(data=X_chilli)
# print(metadata3.to_dict())

metadata4 = SingleTableMetadata()
metadata4.detect_from_dataframe(data=X_Mulberry)
# print(metadata4.to_dict())

metadata5 = SingleTableMetadata()
metadata5.detect_from_dataframe(data=X_Potato)
# print(metadata5.to_dict())

metadata6 = SingleTableMetadata()
metadata6.detect_from_dataframe(data=X_healthy_tomato)
# print(metadata6.to_dict())
```

```python
# Unhealthy_Tomato
synthesizer1 = CTGANSynthesizer(metadata1)
synthesizer1.fit(X_Unhealthy_Tomato)
```

```python
# Capsicum
synthesizer2 = CTGANSynthesizer(metadata2)
synthesizer2.fit(X_Capsicum)
```

```python
# Potato
synthesizer3 = CTGANSynthesizer(metadata3)
synthesizer3.fit(X_Potato)
```

```python
# Chilli
synthesizer4 = CTGANSynthesizer(metadata4)
synthesizer4.fit(X_chilli)
```

```python
# Mulberry
synthesizer5 = CTGANSynthesizer(metadata5)
synthesizer5.fit(X_Mulberry)
```

```python
# healthy_tomato
synthesizer6 = CTGANSynthesizer(metadata6)
synthesizer6.fit(X_healthy_tomato)
```

```python
# Sample generation using trained model
samples_to_gen = 2000
unhealthy_tomato = synthesizer1.sample(num_rows=samples_to_gen)
capsicum         = synthesizer2.sample(num_rows=samples_to_gen)
potato           = synthesizer3.sample(num_rows=samples_to_gen)
chilli           = synthesizer4.sample(num_rows=samples_to_gen)
mulberry         = synthesizer5.sample(num_rows=samples_to_gen)
healthy_tomato   = synthesizer6.sample(num_rows=samples_to_gen)

# Shapes
print("healthy_tomato:", healthy_tomato.shape)
print("Unhealthy_Tomato:", healthy_tomato.shape)
print("Capsicum:", capsicum.shape)
print("Potato:", potato.shape)
print("Chilli:", chilli.shape)
print("Mulberry:", mulberry.shape)
```

```python
# Shuffling

data_chilli= chilli.sample(frac=1)
data_chilli.sample(frac=1)

data_mulberry = mulberry.sample(frac=1)
data_mulberry.sample(frac=1)

data_unhealthy_tomato = unhealthy_tomato.sample(frac=1)
data_unhealthy_tomato.sample(frac=1)

data_healthy_tomato = healthy_tomato.sample(frac=1)
data_healthy_tomato.sample(frac=1)

data_potato = potato.sample(frac=1)
data_potato.sample(frac=1)

data_capsicum = capsicum.sample(frac=1)
data_capsicum.sample(frac=1)
```

```python
# inserting label
data_capsicum.insert(0,'label','capsicum')
data_potato.insert(0,'label',"potato")
data_unhealthy_tomato.insert(0,'label','tomato_unhealthy')
data_healthy_tomato.insert(0,'label','tomato_healthy')
data_mulberry.insert(0,'label','mulberry')
data_chilli.insert(0,'label','chilli')

# saving data
data_capsicum.to_excel("Results/capsicum.xlsx", index = False)
data_healthy_tomato.to_excel("Results/tomato_healthy.xlsx", index = False)
data_unhealthy_tomato.to_excel("Results/tomato_unhealthy.xlsx", index = False)
data_potato.to_excel("Results/potato.xlsx", index = False)
data_mulberry.to_excel("Results/mulberry.xlsx", index = False)
data_chilli.to_excel("Results/chilli.xlsx", index = False)
```
