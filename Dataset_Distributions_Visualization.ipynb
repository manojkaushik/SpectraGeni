```python
from matplotlib import pyplot
import matplotlib.pyplot as plt
import time
from datetime import datetime
import numpy as np
from numpy import mean, isnan, asarray, polyfit
import pandas as pd
from collections import Counter
```

```python
# visulization class distribution

def freq_visualize(resampled_df, title):
    labels = resampled_df['label'].values
    counter = Counter(labels) # counter object
    counter = dict(sorted(counter.items())) # converting counter obj. into dictionary
    x = list(counter.keys())
    y = list(counter.values())
    plt.rcParams.update({'font.size':8})
    fig = plt.figure(num=None, figsize=(10, 4), dpi=200, facecolor='w', edgecolor='red')
    plt.barh(x, y)
    plt.xlim(0, 3000)
    plt.ylabel("Labels", weight ='bold')
    plt.xlabel("Frequencies", weight ='bold')
    plt.title(title)
    plt.show()
```

```python
# Plotting for original data
path = r"C:\Users\IIST\Desktop\SRINIJA\Data\3.Merged_Data\merge.xlsx"
data_Original_crops = pd.read_excel(path, index_col=None, usecols=["label"])
freq_visualize(data_Original_crops, title="Crops Frequencies in Original dataset")
```

```python
# Plotting for Synthetic SMOTE data
path = r"C:\Users\IIST\Desktop\SRINIJA\Data\5.Augmented_SMOTE_Data\7.Combined_Augmented_SMOTE_Data.xlsx"
data_Synthetic_SMOTE_crops = pd.read_excel(path, index_col=None, usecols=["label"])
freq_visualize(data_Synthetic_SMOTE_crops, title="Crops Frequencies in Synthetic_SMOTE dataset")
```

```python
# Plotting for Synthetic Borderline_SMOTE data
path = r"C:\Users\IIST\Desktop\SRINIJA\Data\5.Augmented_SMOTE_Data\....xlsx"
data_Synthetic_Borderline_SMOTE_crops = pd.read_excel(path, index_col=None, usecols=["label"])
freq_visualize(data_Synthetic_Borderline_SMOTE_crops, title="Crops Frequencies in Synthetic_Borderline_SMOTE dataset")
```
